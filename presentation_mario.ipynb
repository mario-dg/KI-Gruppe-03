```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
def to_date_df(df):
    # Check if 'Date' column is already in datetime format
    if not pd.api.types.is_datetime64_any_dtype(df['date']):
        # Convert 'Date' column to datetime type
        df['date'] = pd.to_datetime(df['date'])
    return df.set_index('date')
```

```python
df = pd.read_csv('./data/clean/df_weekly_incidence.csv')
df = to_date_df(df)
df.head()
```

```python
df_hamburg = df[df['administrative_area_level_3'].str.contains("Hamburg")]
display(df_hamburg)
```

```python
df_hamburg = df_hamburg[df_hamburg.index != "2021-01-02 00:00:00"]
display(df_hamburg)
```

```python
hamburg_incidence = list(df_hamburg['incidence'])
hamburg_incidence
```

```python
def plot_simple_time_series(df, column_names):
    plt.style.use('seaborn-whitegrid')
    plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size as needed
    plt.plot(df.index, df[column_names])  # Replace 'value' with your actual data column name
    plt.title('7-Tage Inzidenz der Coronainfektionen (COVID-19) in Hamburg 03.2020-11.2022')
    plt.xlabel('Datum')
    plt.ylabel('7-Tage Inzidenz')
    plt.grid(True)
    plt.show()
```

```python
plot_simple_time_series(df_hamburg, ['incidence'])
```
